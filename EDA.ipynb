<h5><p>let's visualize and explore our data</p></h5>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
df = pd.read_csv('./clean_anime_dataframe')
```

```python
df.corr()
```

<p>From looking at the correlation between all columns on the dataframe, you can see Mean Score is most influenced by Popularity and Favorites, although some genres have more weight than others, most of them are neglible.</p>
<BR>
    <p>Another interesting observation is that the correlation between Popularity and Favorites is a whopping 0.863, which means that on animes with a lot of popularity a lot of people also favorited them.</p>

```python
df[['Mean Score', 'Popularity', 'Favorites']].corr()
```

<p>I still want to see how each Source fairs with the average Mean Score</p>

```python
df.groupby('Source')['Mean Score'].mean().plot(kind = 'bar')
plt.ylabel('Mean Score')
```

<p>Looks like no Source has a definite advantage over the other when it comes to the score of the anime, although 7 which correlates to "Web Novel" seems to have a slightly higher average and 9 which correlates to "Picture book" has a significantly lower score than its counterparts, making it seem like the most popular Animes are taken from web novels.</p>

<div>Now how about the Format? is there a format more popular than others?</div>

```python
fig = plt.figure(figsize = (20,5)) ##longer due to more samples
df.groupby('Format')['Mean Score'].mean().plot(kind = 'bar')
plt.ylabel('Mean Score')
```

<p>Again, seems like most Formats have around the same average score (around 60) but the outliers are more extreme this time, Format 31 that correlates to "ONA(Taiwanese)" is standing above the others at 80 score average, and Format 30 that correlates to "OVA(Chinese)" is way below the rest at around 30 average score. </p>

<p>Now I'd like to see how "Popularity" and "Favorites" affect the score</p>

```python
score_affected = df.groupby(['Mean Score']).mean()[['Popularity', 'Favorites']]
score_affected.plot()
```

<p>As previously seen in the correlation table, the line plot just visualized that the most popular anime get the highest scores.
<BR>
And surprisingly most Anime did not have anyone mark them in favorites, and those who did get marked were in the higher score range, somewhere between 80 and 100.</p>

<p>Now I'd like to see the spread of the Mean Score, what's most common and likely to predict?</p>

```python
fig = plt.figure(figsize = (10,5))
df['Mean Score'].hist(bins = 100, grid = False, density = True)
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
```

<p>Looks like most animes range between 50 to 80 with some outliers close to the 90+ score and the 20's.
</p>
    <p>from this I can speculate that the model I'll use for predicting the score will most likely predict around the range of the most frequent scores.</p>

<p>Now I'd like to see if I can a pattern in which I could predict the score, for that I'll use a scatter plot.</p>

```python
fig, axs = plt.subplots(3, figsize = (20,20))
fig.suptitle('Different scaling to most correlated columns')
##Not resized
axs[0].scatter(x=df['Mean Score'],y=df['Popularity'],c='r',marker='*')
axs[0].scatter(x=df['Mean Score'],y=df['Favorites'],c='b',marker='o')
##Square root
axs[1].scatter(x=df['Mean Score'],y=df['Popularity']**(1/2),c='r',marker='*')
axs[1].scatter(x=df['Mean Score'],y=df['Favorites']**(1/2),c='b',marker='o')
##Lan 
axs[2].scatter(x=df['Mean Score'],y=np.log(df['Popularity']),c='r',marker='*')
axs[2].scatter(x=df['Mean Score'],y=np.log(df['Favorites']),c='b',marker='o')
```

<p>Looking at these 3 subplots, I can tell that the one most likely to work with a Linear regression type model is the 3rd one, the spread of the values are stacked in a much more linear way compared to the first one(which looks like a non linear expotential curve), and slightly more than the 2nd one.

Considering this, I'd most likely use the Lan scaled down version in my prediction model.</p>
